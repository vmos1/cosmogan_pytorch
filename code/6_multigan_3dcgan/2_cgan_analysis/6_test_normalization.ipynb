```python
### Transformation functions for image pixel values
def f_transform(x,a):
    return 2.*x/(x + float(a)) - 1.

def f_invtransform(s,a):
    return float(a)*(1. + s)/(1. - s)
```

$ s = \frac{ 2 \ x } {x + a } -1 $

```python

bins=np.arange(0,2000)
```

```python
a1=np.arange(0,1000)
bins=np.concatenate([np.array([-0.5,1.0]),np.arange(0.5,100.5,5),np.arange(100.5,300.5,20),np.arange(300.5,1000.5,50),np.array([2000])])
bins
```

```python
plt.figure()
for kappa in [2,4,10,20,50,100]:
    y1=f_transform(bins,kappa)
    plt.plot(bins,y1,marker='o',linestyle='',label=kappa)

plt.legend(loc='best')
plt.xscale('log')
plt.show()
```

$$ ρ_{DM} = \log_{10}(1 + ρ_{DM})/12 $$

```python
def f_transf_log(x,a=12.0):
    return np.log10(1+x)/a

def f_invtransf_log(s,a=12.0):
    return 10**(a*s) -1

```

```python
for x in [0,1,10,100,1000,3000]:
    y=f_transf_log(x)
    x2=f_invtransf_log(y)
    print(x,y,x2)
```

```python
plt.figure()
y=f_transf_log(bins,12)
plt.plot(bins,y,marker='o',linestyle='',label='log norm')
# plt.xscale('log')
plt.show()
```

```python
plt.figure()
y=f_transf_log(bins,0.5)
y2=f_transform(bins,50)
plt.plot(bins,y,marker='o',linestyle='',label='log norm')
plt.plot(bins,y2,marker='*',linestyle='',label='old norm')
plt.xscale('log')
plt.legend()
plt.show()
```
